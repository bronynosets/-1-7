```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import networkx as nx
```

```python
G = nx.read_edgelist("C:\\Users\\lesha\\Documents\\Большие графы\\graph.txt")
```

```python
from collections import defaultdict
degrees = defaultdict(int)

for v in G.nodes():
    d = G.degree(v)
    degrees[d] += 1
```

```python
x = sorted(degrees.keys())
y = [degrees[d] for d in x]
```

```python
plt.loglog(x, y, ls='None', marker='.')
plt.show()
```

```python
cd = defaultdict(int)
s = 0
for d in sorted(degrees.keys(), reverse=True):
    s += degrees[d]
    cd[d] = s
```

```python
x1 = sorted(cd.keys())
y1 = [cd[d] for d in x]
plt.loglog(x1, y1, ls='None', marker='.')
plt.show()
```

```python
from scipy.optimize import curve_fit
def func(x, a, c):
    return c/(x**a)
cd1 = defaultdict(int)
for d in sorted(cd.keys(), reverse=True):
    if(cd[d] <= 100000):
        cd1[d] = cd[d]
xdata = sorted(cd.keys())
ydata = [cd[d] for d in x]
plt.loglog(xdata, ydata, ls='None', marker='.')
xdata = sorted(cd1.keys())
ydata = [cd1[d] for d in xdata]
popt, pcov = curve_fit(func, xdata, ydata)
print(popt)
plt.loglog(xdata, func(xdata, *popt))
#popt
plt.show()
```
